```python
#import all libraries that are needed
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the dataset using the path
terrorism = pd.read_csv('/content/gdrive/MyDrive/terrorism/globalterrorismdb_0718dist(3).csv',encoding='latin1')
```

```python
# get first five rows of the terrorism dataset.
terrorism.head()
```

```python
# Gives count of no. of rows and no.of columns
terrorism .shape
```

```python
#describes the dataset
terrorism.describe
```

```python
#checks whether the dataset is having null values or missing values or not.
terrorism.isnull()
```

```python
#this tail function is used to get the last rows of the dataset.
terrorism.tail()
```

```python
#Rename the columns that are required.
terrorism =terrorism.rename(columns={'iyear':'year','imonth':'month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType',
                          'target1':'Target','nkill':'Killed','nwound':'wounded','Summary':'summary','gname':'Group','targtype1_txt':'Target_Type',
                          'weaptype1_txt':'weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'})
```

```python
terrorism['casualities'] = terrorism.Killed + terrorism.wounded
```

```python
#delete all the unwanted columns to simplify the process.
terrorism = terrorism[['year','month','Day','Country','Region','AttackType','Target','Killed','wounded','summary','Group','Target_Type','weapon_type','Motive','City','Latitude','Longitude','casualities']]
```

```python
#Get the first five rows of the dataset after modifying and deleting the unwanted columns.
terrorism.head()
```

```python
#gives the  total  number of missing values in the dataset.
terrorism.isnull().sum()
```

```python
#countplot to describe number of attacks each year
plt.subplots(figsize=(14,7))
sns.countplot('year',data=terrorism,palette='rainbow',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=90)
plt.title('Number of  Terrorist Attacks each year')
plt.show()
```

```python
#barplot to represent causualities in each year.
plt.subplots(figsize=(14,7))
y_casualities= terrorism.groupby('year').casualities.sum().to_frame().reset_index()
y_casualities.columns=['year','casualities']
sns.barplot(x=y_casualities.year,y=y_casualities.casualities,palette='rainbow')
plt.xticks(rotation=90)
plt.title('Number of casualities each year')
plt.show()
```

```python
#barplot representing no. of attacks in each country
plt.subplots(figsize=(14,7))
attacks_country=terrorism.Country.value_counts()[:15].reset_index()
attacks_country.columns = ['Country', 'Total Attacks']
sns.barplot(x=attacks_country.Country, y=attacks_country['Total Attacks'], palette='RdYlGn_r')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('count')
plt.title('Number of attacks in each Country')
plt.show()
```

```python
#Gives exact valuues of attacks in each country
print(terrorism['Country'].value_counts().head())
```

```python
#barplot representing hot regions of terrorist attacks
plt.figure(figsize = (10,5))
sns.barplot(terrorism['Region'].value_counts()[:15].index,terrorism['Region'].value_counts()[:15].values,palette='OrRd_r')
plt.title('Region with high terrorist attacks')
plt.xticks(rotation=90)
plt.xlabel('Region')
plt.ylabel('count')
plt.show()
```

```python
print(terrorism['Region'].value_counts().head())
```

```python
#countplot representing types of weapons that are used
plt.subplots(figsize=(10,7))
sns.countplot('weapon_type',data=terrorism,palette='rainbow')
plt.xticks(rotation=90)
plt.title('Weapons that are used most')
plt.xlabel('weapon types')
plt.ylabel('count')
plt.show()
```

```python
#Gives accurate values of number of weapons used and its types.
print(terrorism['weapon_type'].value_counts().head())
```
